# Data Import

```python
# School profile endpoints
school_profiles= {
    "school_profile_1819": "https://data.cityofchicago.org/resource/kh4r-387c.json",
    # school_profile_1920: "No record"
   "school_profile_2021": "https://data.cityofchicago.org/resource/83yd-jxxw.json",
    "school_profile_2122": "https://data.cityofchicago.org/resource/2dem-8rq7.json",
    "school_profile_2223": "https://data.cityofchicago.org/resource/9a5f-2r4p.json"
}
```

```python
school_profile_df = pd.DataFrame()

for profile in school_profiles:
    url = school_profiles[profile]
    df = pd.DataFrame(requests.get(url).json())
    df["file_name"] = profile
    df.columns = [c.lower() for c in df.columns]
    school_profile_df = school_profile_df.append(df)
```

```python
current_profile = school_profile_df[school_profile_df["file_name"]=='school_profile_2223']
```
